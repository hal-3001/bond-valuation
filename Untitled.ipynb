```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import scipy.stats as stat
```

# creating hypothetical portfolio with equal weightages of tcs,wipro,infosys


##### In the below box i have taken the 5 years stock data of  tcs,wipro,infosys nifty_it,nifty_50

```python
infosys=pd.read_csv("INFY.NS.csv").set_index("Date")
tcs=pd.read_csv("TCS.NS.csv").set_index("Date")
wipro=pd.read_csv("WIPRO.NS.csv").set_index("Date")
nifty_50=pd.read_csv("^NSEI.csv").set_index("Date")
nifty_it=pd.read_csv("^CNXIT.csv").set_index("Date")
```

 from the above weekly stock data i have created a hypothetical portfolio called my_portolio with the sum of weekly returns of infosys,wipro, tcs,the reurns are as follows
 

```python
my_columns={"infy":infosys["Adj Close"],"tcs":tcs["Adj Close"],"wipro":wipro["Adj Close"]}
my_portfolio=pd.DataFrame(my_columns).pct_change()
my_portfolio=(my_portfolio*.3333).sum(axis=1)
my_portfolio
```

now to compare the returns of my_portfolio with nifty_50 and nifty_it i have created a table named data with the returns of my_portfolio,nift_50, nifty_it

```python
newcolumns={"portfolio_returns":my_portfolio,
            "nifty_50_returns":nifty_50["Adj Close"].pct_change(),"nifty_it_returns":nifty_it["Adj Close"].pct_change()}
data=pd.DataFrame(newcolumns)
data
```

from the above statistics we can see that my portfolio has greater returns  and  std than the other two indices

# annualized returns

As we have collected weekly returns to compare the returns on such investments with a one-year return, you need to annualize them. The rate of return per year, measured over a period either longer or shorter than a year, is known as the annualized return.

                          annualized return=(1+r)^52-1
                          
the above formula is used while annualizing weekly returns

```python
def annualized_returns(series):
    avg=series.mean()
    value=((1+avg)**260)-1
    return value
print("The annualized return of my portfolio is "+str(annualized_returns(data["portfolio_returns"])))
print("The annualized return of my portfolio is "+str(annualized_returns(data["nifty_50_returns"])))
print("The annualized return of my portfolio is "+str(annualized_returns(data["nifty_it_returns"])))
```

# annualized risk

as we have calculated weekly standard deviation we need to annualize it:

                            annualized risk= weeklystd*sqrt(52)

```python
def annualized_risk(series):
    std=series.std()
    value=((std)*np.sqrt(260))
    return value

print("The annualized risk of my portfolio is "+str(annualized_risk(data["portfolio_returns"])))
print("The annualized risk of nifty_50 is "+str(annualized_risk(data["nifty_50_returns"])))
print("The annualized risk of nifty_it is "+str(annualized_risk(data["nifty_it_returns"])))
  
```

# Sharpe ratio

Sharpe ratio is the measure of risk-adjusted return of a financial portfolio. A portfolio with a higher Sharpe ratio is considered superior relative to its peers.

```python
def sharpe_ratio(returns,risk,riskfree=0.06511):
    value=(returns-riskfree)/risk
    return value
print("sharpe ratio of my portfolio:"+str(sharpe_ratio(annualized_returns(data["portfolio_returns"]),annualized_risk(data["portfolio_returns"]))))
print("sharpe ratio of my portfolio:"+str(sharpe_ratio(annualized_returns(data["nifty_50_returns"]),annualized_risk(data["nifty_50_returns"]))))
print("sharpe ratio of my portfolio:"+str(sharpe_ratio(annualized_returns(data["nifty_it_returns"]),annualized_risk(data["nifty_it_returns"]))))
```
